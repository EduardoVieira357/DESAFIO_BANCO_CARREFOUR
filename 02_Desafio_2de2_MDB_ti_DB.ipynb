### Lendo os dados do MongoDB e criando um Dataframe

```
# LENDO OS DADOS E CRIANDO UM DATAFRAME

df1 = pd.DataFrame(list(collection.find()))
```

```
df1.head()
```

### Iniciando a análise dos dados

#### Selecionando as colunas

```
df1 = df1[['name', 'tweet_volume', 'country','date_time']]
df1.head().sort_values('name')
```

#### Obtendo as primeiras informações

```
df1.info()
```

#### Deletando as linhas com valores de "tweet_volume" == NULL

```
df1 = df1.dropna(subset=['tweet_volume'])
df1.info()
```

#### Checando Valores únicos e repetidos para "name"

```
print('Valores únicos:    ', len(df1['name'].unique()))
print('Valores repetidos: ', len(df1)-len(df1['name'].unique()))
```

#### Criando um Dataframe com valores únicos para "name"

```
df1_uni = df1[~df1.duplicated('name')]
```

#### Ordenando o Dataframe pelo valor do "tweet_volume"

```
df1_uni = df1_uni.sort_values(by=['tweet_volume'], ascending=False)
```

```
df1_uni.info()
```
